```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_results = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
combined_results.head()
```

```python
# Checking the number of mice.
mouse_count = len(combined_results['Mouse ID'].unique())
print(mouse_count)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicateMouse = combined_results[combined_results.duplicated(['Mouse ID','Timepoint'])]
duplicateMouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicateDF = combined_results[combined_results['Mouse ID'] == 'g989']
duplicateDF.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse = combined_results[combined_results['Mouse ID'] != 'g989']
clean_mouse.reset_index()
clean_mouse.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_count = len(clean_mouse['Mouse ID'].unique())
print(clean_count)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

tumour_mean_volume = clean_mouse.groupby(['Drug Regimen']).mean()['Tumor Volume (mm3)']
tumor_median_volume = clean_mouse.groupby(['Drug Regimen']).median()['Tumor Volume (mm3)']
tumor_var_volume = clean_mouse.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)']
tumor_std_volume = clean_mouse.groupby(['Drug Regimen']).std()['Tumor Volume (mm3)']
tumor_sem_volume= clean_mouse.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)']

# Assemble the resulting series into a single summary DataFrame.

tumor_volume_summary = pd.DataFrame({"Mean Tumor Volume": tumour_mean_volume, "Median Tumor Volume": tumor_median_volume,
                                    "Variance of Tumor Volume": tumor_var_volume, "St.D of Tumor Volume": tumor_std_volume,
                                     "SEM of Tumor Volume": tumor_sem_volume})
```

```python
tumor_volume_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
clean_mouse.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","std","sem"]})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens: 
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = clean_mouse.groupby(['Mouse ID'])['Timepoint'].max()
max_timepoint = pd.DataFrame(max_timepoint)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
clean_maxtime = pd.merge(clean_mouse, max_timepoint, on=['Mouse ID', 'Timepoint'], how='right')
clean_maxtime
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ['Capomulin','Ramicane','Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
capomulin_data = []
ramicane_data = []
infubinol_data = []
ceftamin_data = []

for index, row in clean_maxtime.iterrows():
    if row['Drug Regimen'] == 'Capomulin':
        capomulin_data.append(row['Tumor Volume (mm3)'])
    elif row['Drug Regimen'] == treatment_list[1]:
        ramicane_data.append(row['Tumor Volume (mm3)'])
    elif row['Drug Regimen'] == treatment_list[2]:
        infubinol_data.append(row['Tumor Volume (mm3)'])
    elif row['Drug Regimen'] == treatment_list[3]:
        ceftamin_data.append(row['Tumor Volume (mm3)'])

drug_regimen_data = {treatment_list[0] : capomulin_data, treatment_list[1]: ramicane_data,
                     treatment_list[2]: infubinol_data , treatment_list[3]: ceftamin_data}
drug_regimen_data_df = pd.DataFrame(drug_regimen_data)

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
cap_quart = drug_regimen_data_df[treatment_list[0]].quantile([0.25, 0.5, 0.75])
ram_quart = drug_regimen_data_df[treatment_list[1]].quantile([0.25, 0.5, 0.75])
inf_quart = drug_regimen_data_df[treatment_list[2]].quantile([0.25, 0.5, 0.75])
cef_quart = drug_regimen_data_df[treatment_list[3]].quantile([0.25, 0.5, 0.75])
drug_regimen_data_df
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds
```

```python
cap_iqr = cap_quart[0.75] - cap_quart[0.25]
ram_iqr = ram_quart[0.75] - ram_quart[0.25]
inf_iqr = inf_quart[0.75] - inf_quart[0.25]
cef_iqr = cef_quart[0.75] - cef_quart[0.25]
```

```python
cap_upper = cap_quart[0.75] + (cap_iqr*1.5)
cap_lower = cap_quart[0.25] - (cap_iqr*1.5)
max_cap = drug_regimen_data_df[treatment_list[0]].max()
min_cap = drug_regimen_data_df[treatment_list[0]].min()

ram_upper = ram_quart[0.75] + (ram_iqr*1.5)
ram_lower = ram_quart[0.25] - (ram_iqr*1.5)
max_ram = drug_regimen_data_df[treatment_list[1]].max()
min_ram = drug_regimen_data_df[treatment_list[1]].min()

inf_upper = inf_quart[0.75] + (inf_iqr*1.5)
inf_lower = inf_quart[0.25] - (inf_iqr*1.5)
max_inf = drug_regimen_data_df[treatment_list[2]].max()
min_inf = drug_regimen_data_df[treatment_list[2]].min()

cef_upper = cef_quart[0.75] + (cef_iqr*1.5)
cef_lower = cef_quart[0.25] - (cef_iqr*1.5)
max_cef = drug_regimen_data_df[treatment_list[3]].max()
min_cef = drug_regimen_data_df[treatment_list[3]].min()

drug_regimen_analysis = {'Drug Regimen': [treatment_list[0], treatment_list[1], treatment_list[2], treatment_list[3]], 
                         'Upper Quartile': [cap_quart[0.75], ram_quart[0.75 ], inf_quart[0.75], cef_quart[0.75]], 
                         'Lower Quartile': [cap_quart[0.25], ram_quart[0.25 ], inf_quart[0.25], cef_quart[0.25]], 
                        'Upper Bound': [cap_upper, ram_upper, inf_upper, cef_upper], 'Lower Bound': [cap_lower, ram_lower, inf_lower, cef_lower], 
                         'IQR': [cap_iqr, ram_iqr, inf_iqr, cef_iqr], 'Max Tumor Vol': [max_cap, max_ram, max_inf, max_cef], 
                         'Min Tumor Vol': [min_cap, min_ram, min_inf, min_cef]}

drug_regimen_analysis_df = pd.DataFrame(drug_regimen_analysis)
drug_regimen_analysis_df
```

```python
print(f'{treatment_list[2]} has an outlier value of {min_inf}. It is below the lower bound of {inf_lower}.')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
drug_fig, ax1 = plt.subplots()
ax1.set_title('Distrubution of Tumor Volume by Drug Regimen')
ax1.set_ylabel('Tumor Volume (mm3)')
data = [capomulin_data, ramicane_data, infubinol_data, ceftamin_data]
ax1.boxplot(data, labels=treatment_list)
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
cap_df = clean_mouse.loc[clean_mouse["Drug Regimen"] == "Capomulin",:]
cap_avg = cap_df.groupby(['Mouse ID']).mean()
plt.scatter(cap_avg['Weight (g)'], cap_avg['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Avg. Tumor Volume (mm3)')
plt.title('Weight vs. Avg. Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
cap_corr = round(st.pearsonr(cap_avg['Weight (g)'],cap_avg['Tumor Volume (mm3)'])[0],2)
# for mouse weight and average tumor volume for the Capomulin regimen
cap_slope, cap_y_int, cap_r_val, cap_p_val, cap_std_err = st.linregress(cap_avg['Weight (g)'], cap_avg['Tumor Volume (mm3)'])
cap_line_equ = f'y = {cap_slope}x + {cap_y_int}'
cap_fit = cap_avg['Weight (g)']*cap_slope + cap_y_int
print(f'Corr. Coeficient: {cap_corr}')
```

```python
plt.scatter(cap_avg['Weight (g)'], cap_avg['Tumor Volume (mm3)'], color='yellow')
plt.plot(cap_avg['Weight (g)'], cap_fit, color='red', label=cap_line_equ)
plt.xlabel('Weight (g)')
plt.ylabel('Avg. Tumor Volume (mm3)')
plt.title('Weight vs. Avg. Tumor Volume (mm3)')
plt.legend()
plt.show()
```
